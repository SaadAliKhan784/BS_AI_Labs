```python

# Student Result Analysis using NumPy
import numpy as np

data = np.array([
    [78, 85, 80, 90, 1],
    [45, 50, 40, 55, 0],
    [18, 72, 94, 95, 1],
    [50, 95, 40, 58, 1],
    [75, 70, 68, 52, 0],
    [80, 35, 38, 91, 1],
    [25, 78, 90, 72, 1],
    [98, 15, 50, 87, 0],
    [42, 55, 88, 80, 1],
    [60, 25, 88, 60, 0]
])


X = data[:, :-1]
y = data[:, -1]

print("Features (X):")
print(X)

print("\nLabels (y):")
print(y)


print("\nAverage Marks:")
print(np.mean(X, axis=0))

print("\nMaximum Marks:")
print(np.max(X, axis=0))

print("\nMinimum Marks:")
print(np.min(X, axis=0))


passed = np.sum(y == 1)
failed = np.sum(y == 0)

print("\nNumber of Passed Students:", passed)
print("Number of Failed Students:", failed)


above_80 = X[X > 80]

print("\nMarks Above 80:")
print(above_80)


below_50 = X[X < 50]

print("\nMarks Below 50:")
print(below_50)

X_normalized = (X - np.min(X, axis=0)) / (
    np.max(X, axis=0) - np.min(X, axis=0)
)

print("\nNormalized Dataset:")
print(X_normalized)

average_marks = np.mean(X, axis=1)

predictions = np.where(average_marks >= 50, 1, 0)

print("\nPredicted Results:")
print(predictions)

accuracy = np.mean(predictions == y) * 100

print("\nAccuracy of Predictions:")
print(accuracy, "%")
```
