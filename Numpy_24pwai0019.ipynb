```python
import numpy as np
arr1 = np.arange(1, 11)        # 1D array 1–10
arr2 = np.arange(1, 10).reshape(3, 3)  # 3x3 array
zeros = np.zeros((4, 4))
ones = np.ones((2, 5))

print(arr1)
print(arr2)
print(zeros)
print(ones)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Shape:", arr.shape)
print("Total elements:", arr.size)
print("Data type:", arr.dtype)
print("Dimensions:", arr.ndim)
```

```python
arr = np.array([10, 20, 30, 40, 50, 60])

print("First:", arr[0])
print("Last:", arr[-1])
print("Third:", arr[2])

print("Greater than 30:", arr[arr > 30])
```

```python
arr = np.array([10, 20, 30, 40, 50, 60])

print(arr[1:5])
print(arr[::-1])
print(arr[::2])

matrix = np.arange(1, 10).reshape(3, 3)

print("First row:", matrix[0])
print("Second column:", matrix[:, 1])
print("Last two rows:", matrix[1:])
```

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr * 5)
print(arr + 10)
print(arr ** 2)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr + 10)
print(arr * 2)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Addition:\n", A + B)
print("Subtraction:\n", A - B)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Dot Product:\n", np.dot(A, B))
print("Transpose:\n", A.T)
```

```python
rand_int = np.random.randint(1, 101, (4, 4))
rand_float = np.random.rand(4, 4)

print(rand_int)
print(rand_float)

print("Max:", rand_int.max())
print("Min:", rand_int.min())
print("Average:", rand_int.mean())
```

```python
marks = np.array([78, 85, 90, 45, 60, 30, 88, 92, 55, 70])

print("Average:", marks.mean())
print("Highest:", marks.max())
print("Lowest:", marks.min())

print("Above 80:", np.sum(marks > 80))
print("Failed (<50):", marks[marks < 50])
```

```python
img = np.random.randint(0, 256, (5, 5))

print(img)
print("Brightest:", img.max())
print("Darkest:", img.min())
print("Shape:", img.shape)
```

```python
data = np.array([10, 20, 30, 40, 50])

normalized = (data - data.min()) / (data.max() - data.min())

print(normalized)
```

```python
# Age, Salary, Purchased
data = np.array([
    [25, 50000, 0],
    [30, 60000, 1],
    [35, 65000, 1],
    [40, 70000, 0],
    [28, 52000, 1]
])

X = data[:, :2]   # Features
y = data[:, 2]    # Labels

print("Features:\n", X)
print("Labels:\n", y)

print("Average Salary:", np.mean(data[:, 1]))
print("Highest Age:", np.max(data[:, 0]))
print("Shape:", data.shape)
```
