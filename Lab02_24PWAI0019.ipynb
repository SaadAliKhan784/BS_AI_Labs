```python
# Task 02: Simple calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))


operation = input("Enter operation (+, -, *, /, **, %): ")


if operation not in ["+", "-", "*", "/", "**", "%"]:
    print("Invalid operator!")
else:

    if operation == "/" and num2 == 0:
        print("Cannot divide by zero!")
    else:

        if operation == "+":
            result = num1 + num2
            print(f"{num1} + {num2} = {result}")
        elif operation == "-":
            result = num1 - num2
            print(f"{num1} - {num2} = {result}")
        elif operation == "*":
            result = num1 * num2
            print(f"{num1} * {num2} = {result}")
        elif operation == "/":
            result = num1 / num2
            print(f"{num1} / {num2} = {result:.2f}")
        elif operation == "**":
            result = num1 ** num2
            print(f"{num1} ** {num2} = {result:.2f}")
        elif operation == "%":
            result = num1 % num2
            print(f"{num1} % {num2} = {result}")
```

```python
# Task 03A: Even/Odd & Multiple Checker

start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

num = start
while num <= end:

    if num % 7 == 0:
        num += 1
        continue


    if num % 2 == 0 and num % 5 == 0:
        print(f"{num}: Even & Multiple of 5!!")
    elif num % 2 == 0:
        print(f"{num}: Even")
    elif num % 5 == 0:
        print(f"{num}: Multiple of 5!")
    else:
        print(num)

    num += 1
```

```python
# Task 03B: Password Strength Checker

password = input("Enter a password: ")


if len(password) < 6:
    print("Too short!")
else:

    if not any(char.isdigit() for char in password):
        print("Add at least one number")
    else:

        if not any(char.isupper() for char in password):
            print("Add at least one capital letter")
        else:
            print("Strong password – good choice!")
```
