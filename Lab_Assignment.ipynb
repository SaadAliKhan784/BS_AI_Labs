```python
import numpy as np

# STUDENT DATASET
data = np.array([
    [78, 85, 80, 90, 1],
    [45, 50, 40, 55, 0],
    [88, 92, 81, 95, 1],
    [60, 65, 58, 62, 1],
    [35, 40, 30, 45, 0],
    [95, 98, 96, 99, 1],
    [55, 52, 60, 58, 1],
    [48, 45, 50, 42, 0],
    [82, 78, 85, 88, 1],
    [70, 72, 68, 75, 1]
])

print("Dataset:\n")
print(data)
```

```python
# SEPARATE FEATURES (X) AND LABELS (y)

X = data[:, 0:4]
y = data[:, 4]

print("\nFeatures (X):\n")
print(X)

print("\nLabels (y):\n")
print(y)
```

```python
# AVERAGE, MAXIMUM, AND MINIMUM MARKS

average_marks = np.mean(X, axis=0)
maximum_marks = np.max(X, axis=0)
minimum_marks = np.min(X, axis=0)

subjects = ["Math", "Science", "English", "Computer"]

print("\nAverage Marks:")
for i in range(4):
    print(subjects[i], "=", average_marks[i])

print("\nMaximum Marks:")
for i in range(4):
    print(subjects[i], "=", maximum_marks[i])

print("\nMinimum Marks:")
for i in range(4):
    print(subjects[i], "=", minimum_marks[i])
```

```python
# NUMBER OF PASSED AND FAILED STUDENTS

passed = np.sum(y == 1)
failed = np.sum(y == 0)

print("\nNumber of Passed Students:", passed)
print("Number of Failed Students:", failed)
```

```python
# STUDENTS SCORING ABOVE 80 AND BELOW 50

above_80 = X[X > 80]
below_50 = X[X < 50]

print("\nMarks Above 80:")
print(above_80)

print("\nMarks Below 50:")
print(below_50)
```

```python
# NORMALIZATION OF DATASET
# Formula:
# (X - min) / (max - min)
X_normalized = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

print("\nNormalized Dataset:\n")
print(X_normalized)
```

```python
# SIMPLE PREDICTION LOGIC

average_student_marks = np.mean(X, axis=1)

predictions = np.where(average_student_marks >= 60, 1, 0)

print("\nPredictions:")
print(predictions)
```

```python
# ACCURACY CALCULATION
accuracy = np.mean(predictions == y) * 100

print("\nAccuracy of Predictions:", accuracy, "%")
```
